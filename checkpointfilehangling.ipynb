```python
def read_entire_file(filename):
    with open(filename, 'r') as file:
        content = file.read()
    return content

def read_first_n_lines(filename, n):
    with open(filename, 'r') as file:
        lines = [next(file) for _ in range(n)]
    return ''.join(lines)

def read_last_n_lines(filename, n):
    with open(filename, 'r') as file:
        lines = file.readlines()
        last_n_lines = ''.join(lines[-n:])
    return last_n_lines

def count_words(filename):
    with open(filename, 'r') as file:
        content = file.read()
        words = content.split()
    return len(words)

def read_last_n_lines_bonus(filename, n):
    with open(filename, 'r') as file:
        lines = file.readlines()
        last_n_lines = ''.join(lines[-n:])
    return last_n_lines

filename = 'your_file.txt'
n_first = 5  # Number of lines to read for the first task
n_last = 3   # Number of lines to read for the third task
n_bonus = 4  # Number of lines to read for the bonus task

# Task 1
file_content = read_entire_file(filename)
print("Entire file content:")
print(file_content)

# Task 2
first_n_lines = read_first_n_lines(filename, n_first)
print("\nFirst", n_first, "lines:")
print(first_n_lines)

# Task 3
last_n_lines = read_last_n_lines(filename, n_last)
print("\nLast", n_last, "lines:")
print(last_n_lines)

# Task 4
word_count = count_words(filename)
print("\nNumber of words:", word_count)

# Bonus Task
last_n_lines_bonus = read_last_n_lines_bonus(filename, n_bonus)
print("\nBonus - Last", n_bonus, "lines:")
print(last_n_lines_bonus)
```
